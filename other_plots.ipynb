```python
import numpy as np
from scipy.interpolate import griddata

fig = plt.figure(figsize=(6,4))
plt.tick_params(which = 'major', length = 5, direction='in', bottom=True, top = True, left = True, right=True)
plt.tick_params(labelbottom=True, labeltop=False, labelleft=True,labelright=False)
xi = np.linspace(df_part_2_10['N1'].min(),df_part_2_10['N1'].max(), 50)
yi = np.linspace(df_part_2_10['C2'].min(),df_part_2_10['C2'].max(), 50)
xi,yi = np.meshgrid(xi,yi)

zi = griddata((df_part_2_10['N1'],df_part_2_10['C2']), df_part_2_10['r'], (xi,yi), method = 'cubic')

zi = np.nan_to_num(zi, nan=0.0)

contour = plt.contour(yi,xi,zi, levels = [1.,zi.max()], colors = 'black', alpha = 0.8, linewidth=3)
contour.collections[0].set_label('obs. exclusion')
plt.scatter(df_part_2_10['C2'], df_part_2_10['N1'], c = df_part_2_10['r'], cmap = 'jet')
plt.clabel(contour, inline=True, fmt=r'$r_{\rm max} =$ '+'%i')
cbar = plt.colorbar()
cbar.set_label(r'$r_{max}$')

plt.text(300,480,r'$M_1 = 2000$ GeV'+'\n' +r'$\tan\beta = 10$' '\n' r'$M_2 \geq \mu$', fontsize = 10, bbox=dict(facecolor='skyblue', alpha = 0.5, edgecolor='black', boxstyle='round,pad=0.5'))
#plt.text(300,525,r'$\tan\beta = 10$', fontsize = 12)
#plt.text(300,500,r'$M_2 \geq \mu$', fontsize = 12)
plt.grid(alpha=0.6)
plt.xlabel(r'$m_{\tilde\chi_2^\pm}$ [GeV]', fontsize = 14)
plt.ylabel(r'$m_{\tilde\chi_1^0}$ [GeV]', fontsize = 14)
plt.title(r'exclusion map with any $r_{exp}$, minmassgap=10 GeV')
plt.tight_layout()
plt.savefig('plot_N1_C2_new.png')
```
