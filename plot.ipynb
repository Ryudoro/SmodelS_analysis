```python

import numpy as np
from scipy.interpolate import griddata

xi = np.linspace(max_r_values_df['mu'].min(),max_r_values_df['mu'].max(), 50)
yi = np.linspace(max_r_values_df['m2'].min(),max_r_values_df['m2'].max(), 50)
xi,yi = np.meshgrid(xi,yi)

zi = griddata((max_r_values_df['mu'],max_r_values_df['m2']), max_r_values_df['r'], (xi,yi), method = 'cubic')

plt.scatter(max_r_values_df['mu'], max_r_values_df['m2'], c = max_r_values_df['r'])
zi = np.nan_to_num(zi, nan=0.0)
plt.contourf(xi,yi,zi, levels = np.arange(1,zi.max(),0.1), cmap='plasma', alpha = 0.8)
plt.xlabel('mu')
plt.ylabel('m2')
plt.title('Contour plot for rmax > 1')
plt.savefig('contour_plot.png')
```
