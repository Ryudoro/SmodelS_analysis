```python
import matplotlib.pyplot as plt

from extract_data import main_extract
from extract_input import extract_m1_m2_mu
from extract_input import extract_N1_N2_C1
```

```python
#Importation des dataframes bruts
df_20,liste_input_20 = main_extract('output_minmassgap_20')
df_10, liste_input_10 = main_extract('output_minmassgap_10')
df_5, liste_input_5 = main_extract('output_minmassgap_5')
```

```python
#Ajout d'informations
df_20['m1'], df_20['m2'], df_20['mu'] = zip(*df_20['input_param'].apply(extract_m1_m2_mu))
df_20['N1'], df_20['N2'], df_20['C1'], df_20['C2'] = zip(*df_20['input_param'].apply(extract_N1_N2_C1))

max_r_index_20 = df_20.groupby('input_param')['r'].idxmax()
max_r_expected_index_20= df_20.groupby('input_param')['r_expected'].idxmax()

max_r_values_df_20 = df_20.loc[max_r_index_20]
max_r_expected_values_df_20 = df_20.loc[max_r_expected_index_20]

df_10['m1'], df_10['m2'], df_10['mu'] = zip(*df_10['input_param'].apply(extract_m1_m2_mu))
df_10['N1'], df_10['N2'], df_10['C1'], df_10['C2'] = zip(*df_10['input_param'].apply(extract_N1_N2_C1))

max_r_index_10 = df_10.groupby('input_param')['r'].idxmax()
max_r_expected_index_10 = df_10.groupby('input_param')['r_expected'].idxmax()

max_r_values_df_10 = df_10.loc[max_r_index_10]
max_r_expected_values_df_10 = df_10.loc[max_r_expected_index_10]

df_5['m1'], df_5['m2'], df_5['mu'] = zip(*df_5['input_param'].apply(extract_m1_m2_mu))
df_5['N1'], df_5['N2'], df_5['C1'], df_5['C2'] = zip(*df_5['input_param'].apply(extract_N1_N2_C1))

max_r_index_5 = df_5.groupby('input_param')['r'].idxmax()
max_r_expected_index_5 = df_5.groupby('input_param')['r_expected'].idxmax()

max_r_values_df_5 = df_5.loc[max_r_index_5]
max_r_expected_values_df_5 = df_5.loc[max_r_expected_index_5]


max_r_values_df_10.head()
```

```python
#Séparations pour plot (N1,C2)
df_part_1_5 = max_r_expected_values_df_5[max_r_expected_values_df_5['mu']>max_r_expected_values_df_5['m2']]
df_part_2_5 = max_r_expected_values_df_5[max_r_expected_values_df_5['mu']<max_r_expected_values_df_5['m2']]

df_part_1_10 = max_r_expected_values_df_10[max_r_expected_values_df_10['mu']>max_r_expected_values_df_10['m2']]
df_part_2_10 = max_r_expected_values_df_10[max_r_expected_values_df_10['mu']<max_r_expected_values_df_10['m2']]

df_part_1_20 = max_r_expected_values_df_20[max_r_expected_values_df_20['mu']>max_r_expected_values_df_20['m2']]
df_part_2_20 = max_r_expected_values_df_20[max_r_expected_values_df_20['mu']<max_r_expected_values_df_20['m2']]
```

```python
#Essai de visualiation des différentes valeurs de r (sans préciser les exclusions)
fig, (ax1,ax2 )= plt.subplots(1,2, figsize = (12,6))

import numpy as np

# df_part_1_10['size'] = np.where(df_part_1_10['r'] > 1, 100, 50)
# df_part_2_10['size'] = np.where(df_part_2_10['r'] > 1, 100, 50)

sc2_10 = ax1.scatter(df_part_1_10['N1'], df_part_1_10['C2'], c = df_part_1_10['r'])
sc3_10 = ax2.scatter(df_part_2_10['N1'], df_part_2_10['C2'], c = df_part_2_10['r'])

ax1.set_xlabel('$m_{\tilde\chi^0_1}$')
ax1.set_ylabel('$m_{\tilde\chi^{\pm}_2}$')
ax2.set_xlabel('$m_{\tilde\chi^0_1}$')
ax2.set_ylabel('$m_{\tilde\chi^{\pm}_2}$')

ax1.set_title("$r_{max}$ values for $\mu > M_2$")
ax2.set_title("$r_{max}$ values for $M_2 > \mu$")

plt.colorbar(sc3_10)
```

```python
import numpy as np
from scipy.interpolate import griddata

xi_20 = np.linspace(max_r_expected_values_df_20['mu'].min(),max_r_expected_values_df_20['mu'].max(), 50)
yi_20 = np.linspace(max_r_expected_values_df_20['m2'].min(),max_r_expected_values_df_20['m2'].max(), 50)
xi_20,yi_20 = np.meshgrid(xi_20,yi_20)

zi_20 = griddata((max_r_expected_values_df_20['mu'],max_r_expected_values_df_20['m2']), max_r_expected_values_df_20['r'], (xi_20,yi_20), method = 'cubic')

zi_20 = np.nan_to_num(zi_20, nan=0.0)
plt.contourf(xi_20,yi_20,zi_20, levels = [1,zi_20.max()], colors = 'r', alpha = 0.6)
plt.contour(xi_20,yi_20,zi_20, levels = [1,zi_20.max()], colors = 'r', alpha = 0.6)

xi_10 = np.linspace(max_r_expected_values_df_10['mu'].min(),max_r_expected_values_df_10['mu'].max(), 50)
yi_10 = np.linspace(max_r_expected_values_df_10['m2'].min(),max_r_expected_values_df_10['m2'].max(), 50)
xi_10,yi_10 = np.meshgrid(xi_10,yi_10)

zi_10 = griddata((max_r_expected_values_df_10['mu'],max_r_expected_values_df_10['m2']), max_r_expected_values_df_10['r'], (xi_10,yi_10), method = 'cubic')

zi_10 = np.nan_to_num(zi_10, nan=0.0)
plt.contourf(xi_10,yi_10,zi_10, levels = [1,zi_10.max()], colors = 'orange', alpha = 0.6)
plt.contour(xi_10,yi_10,zi_10, levels = [1,zi_10.max()], colors = 'orange', alpha = 0.8)


xi_5 = np.linspace(max_r_values_df_5['mu'].min(),max_r_values_df_5['mu'].max(), 50)
yi_5 = np.linspace(max_r_values_df_5['m2'].min(),max_r_values_df_5['m2'].max(), 50)
xi_5,yi_5 = np.meshgrid(xi_5,yi_5)

zi_5 = griddata((max_r_values_df_5['mu'],max_r_values_df_5['m2']), max_r_values_df_5['r'], (xi_5,yi_5), method = 'cubic')

zi_5 = np.nan_to_num(zi_5, nan=0.0)
plt.contourf(xi_5,yi_5,zi_5, levels = [1,zi_5.max()], colors = 'blue', alpha = 0.6)
plt.contour(xi_5,yi_5,zi_5, levels = [1,zi_5.max()], colors = 'blue', alpha = 0.8)



plt.scatter(max_r_values_df_20['mu'], max_r_values_df_20['m2'], c = 'black')

plt.xlabel("$\mu$ (GeV)")
plt.ylabel("$M_2$ (GeV)")

plt.title("$r_{max} > 1$ exclusion for minmassgap = 5 GeV (blue), 10 GeV (orange) and 20 GeV (Red)")
plt.savefig("contour_plot_expected_minmassgap_5_10_20_GeV.png")
```

```python
max_r_expected_value_df_5_excluded = max_r_expected_values_df_5[max_r_expected_values_df_5['r'] >= 1]
max_r_expected_value_df_5_not_excluded = max_r_expected_values_df_5[max_r_expected_values_df_5['r'] < 1]
unique_ids = max_r_expected_value_df_5_excluded['AnalysisID'].unique()

colors = plt.cm.viridis(np.linspace(0,1,len(unique_ids)))
color_dict = dict(zip(unique_ids,colors))

for id in unique_ids:
    id_df = max_r_expected_value_df_5_excluded[max_r_expected_value_df_5_excluded['AnalysisID'] == id]
    plt.scatter(id_df['mu'], id_df['m2'], color=color_dict[id], label = id)

plt.scatter(max_r_expected_value_df_5_not_excluded['mu'], max_r_expected_value_df_5_not_excluded['m2'], c = 'gray', alpha = 0.4, label = 'Not excluded')
plt.legend(bbox_to_anchor = (1.04,1.03),loc = 'upper left', frameon= True)

plt.xlabel( '$\mu$ (GeV)')
plt.ylabel('$M_2$ (GeV)')
plt.title('excluded points ($r_{max} > 1$) for minmassgap = 5 GeV')
plt.savefig("scatter_plot_expected_minmassgap=5GeV.png")
```
