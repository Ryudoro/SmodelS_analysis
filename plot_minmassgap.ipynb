```python
from matplotlib.cm import ScalarMappable

max_r_value_df_10_excluded = max_r_values_df_10[max_r_values_df_10['r'] >= 1]
max_r_value_df_10_not_excluded = max_r_values_df_10[max_r_values_df_10['r'] < 1]
unique_ids = max_r_values_df_10['AnalysisID'].unique()

from matplotlib.colors import Normalize
# colors = plt.cm.jet(np.linspace(0,1,len(max_r_values_df_10['r'].unique())))
# color_dict = dict(zip(unique_ids,colors))

# for id in unique_ids:
#     id_df = max_r_value_df_10_excluded[max_r_value_df_10_excluded['AnalysisID'] == id]
#     plt.scatter(id_df['mu'], id_df['m2'], color=color_dict[id], label = id)
markers = ['o', 'v', 's', 'p', '*', '+', 'x', 'd', '1', 'D', 'H', '>']
unique_ids = max_r_values_df_10['AnalysisID'].unique()
print(unique_ids)
marker_dict = dict(zip(unique_ids, markers))

# Normalize r for the color mapping
norm = Normalize(vmin=max_r_values_df_10['r'].min(), vmax=max_r_values_df_10['r'].max())

# Create the figure and axes
fig = plt.figure(figsize=(10,8))
plt.tick_params(which = 'major', length = 5, direction='in', bottom=True, top = True, left = True, right=True)
plt.tick_params(labelbottom=True, labeltop=False, labelleft=True,labelright=False)
xi = np.linspace(max_r_values_df_10['mu'].min(),max_r_values_df_10['mu'].max(), 50)
yi = np.linspace(max_r_values_df_10['m2'].min(),max_r_values_df_10['m2'].max(), 50)
xi,yi = np.meshgrid(xi,yi)

zi = griddata((max_r_values_df_10['mu'],max_r_values_df_10['m2']), max_r_values_df_10['r'], (xi,yi), method = 'cubic')

zi = np.nan_to_num(zi, nan=0.0)

contour = plt.contour(xi,yi,zi, levels = [1.,zi.max()], colors = 'black', alpha = 0.8)
contour.collections[0].set_label('obs. exclusion')
plt.clabel(contour, inline=True, fmt=r'$r_{\rm max} =$ '+'%i')

# Create scatter plots for each AnalysisID
for id in unique_ids:
    id_df = max_r_values_df_10[max_r_values_df_10['AnalysisID'] == id]
    colors = plt.cm.jet(norm(id_df['r']))
    sc = plt.scatter(id_df['mu'], id_df['m2'], c=colors, marker=marker_dict[id], label=id)

#plt.scatter(max_r_values_df_10['mu'], max_r_values_df_10['m2'], c = 'gray', alpha = 0.4, label = 'Not excluded')
plt.legend(bbox_to_anchor = (1.04,1.03),loc = 'upper left', frameon= True)

sm = ScalarMappable(cmap='jet', norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, location='left', pad = 0.1)
cbar.set_label('r')

cbar.set_label(r'$r_{max}$', labelpad=-1)

plt.text(750,1000,r'$M_1 = 2000$ GeV'+'\n' +r'$\tan\beta = 10$' '\n' r'$M_2 \geq \mu$', fontsize = 10, bbox=dict(facecolor='skyblue', alpha = 0.5, edgecolor='black', boxstyle='round,pad=0.5'))

plt.xlabel( r'$\mu$ [GeV]', fontsize = 14)
plt.ylabel(r'$M_2$ [GeV]', fontsize = 14)
plt.title('excluded points ($r_{max} > 1$) for minmassgap = 10 GeV')
plt.grid(alpha=0.6)
plt.tight_layout()
plt.savefig("scatter_plot_minmassgap=10GeV_v2.png")
```
